##### The last column called "fraud" has values of 0s and 1s which suggest 0 is for non-fraud and 1 is for fraud transactions. Lets visualize, create a table the splits the data between fraud and non-fraud transactions, get more insights.


```python
df.pivot_table(index=['fraud'])
```

```python
df.groupby("fraud").mean()
```

#### 1. distance_from_home: 

On average, fraudulent transactions have a higher distance from the home location compared to non-fraudulent transactions. This might suggest that fraudulent transactions tend to occur from locations farther away from the cardholder's registered home address.

#### 2. distance_from_last_transaction: 

The average distance from the last transaction is significantly higher for fraudulent transactions compared to non-fraudulent ones. This could imply that fraudulent transactions often occur at a greater temporal distance from the previous transaction.

#### 3. ratio_to_median_purchase_price: 
The average ratio of the transaction amount to the median purchase price is much higher for fraudulent transactions. This suggests that fraudulent transactions involve higher amounts relative to the typical purchase price, indicating potentially suspicious behavior.

#### 4. repeat_retailer: 

Both fraudulent and non-fraudulent transactions have a similar average value for the "repeat_retailer" feature, which indicates that whether a retailer is repeated in consecutive transactions might not be a strong distinguishing factor between the two groups.

#### 5. used_chip: 

The average value for the "used_chip" feature is lower for fraudulent transactions compared to non-fraudulent ones. This might indicate that a lower proportion of fraudulent transactions involve chip-based card usage, which could be a point of concern.

#### 6. used_pin_number: 

Similarly, the average value for the "used_pin_number" feature is much lower for fraudulent transactions. This could imply that fraudsters tend to avoid using PIN numbers.

#### 7. online_order: 

The average value for the "online_order" feature is substantially higher for fraudulent transactions. This suggests that a higher proportion of fraudulent transactions are conducted online compared to non-fraudulent ones.

We see there is a high correlation between frauded transactions and ratio_to_median_purchase_price. Lets create machine learning models to test the accuracy of these models and if they can predict future frauded transactions.

### Lets dive into this

```python
df.groupby(['fraud'])['ratio_to_median_purchase_price'].mean().plot(kind='bar')
```

We will breakdown the purchases of this column into low, medium and high

```python
df['purchase_category'] = pd.cut(df['ratio_to_median_purchase_price'], bins=[0,3,15,300], labels=['low','medium','high'])
```

```python
sns.countplot(data=df,x='purchase_category')
```

Lets  see how much fraud could we find in these categories

```python
df.groupby('purchase_category')['fraud'].mean().plot(kind='bar')
plt.ylabel('fraud')
```

Lots of fraud happening in the medium and high category, so our model will focus on this column and see if we can create a model that will help us predict and prevent these fraudelant transactions.
